```python
import pandas as pd
import numpy as np



# Define a dictionary of fuel prices, primary energy conversion factors, and emission factors
fuel_data = {
    'mains gas (not community)': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'oil (not community)': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'wood chips': {'price': 0.0581, 'primary_factor': 1.26, 'emission_factor': 0.039},
    'electricity (not community)': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'bottled LPG': {'price': 0.0760, 'primary_factor': 1.09, 'emission_factor': 0.241},
    'LPG (not community)': {'price': 0.0760, 'primary_factor': 1.09, 'emission_factor': 0.241},
    'smokeless coal': {'price': 0.0461, 'primary_factor': 1.21, 'emission_factor': 0.433},
    'house coal (not community)': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'wood logs': {'price': 0.0423, 'primary_factor': 1.04, 'emission_factor': 0.019},
    'anthracite': {'price': 0.0364, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'bulk wood pellets': {'price': 0.0581, 'primary_factor': 1.26, 'emission_factor': 0.039},
    'electricity (community)': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'oil (community)': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'coal (community)': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'biomass (community)': {'price': 0.0581, 'primary_factor': 1.26, 'emission_factor': 0.039},
    'biogas (community)': {'price': 0.0760, 'primary_factor': 1.10, 'emission_factor': 0.098},
    'biodiesel from any biomass source': {'price': 0.0764, 'primary_factor': 1.06, 'emission_factor': 0.123},
    # Add other mappings as needed
    'dual fuel - mineral + wood': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'NO DATA!': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'To be used only when there is no heating/hot-water system': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'Electricity: electricity, unspecified tariff': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'mains gas (community)': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'mains gas - this is for backwards compatibility only and should not be used': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'oil - this is for backwards compatibility only and should not be used': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'LPG - this is for backwards compatibility only and should not be used': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'B30D (community)': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'B30K (not community)': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'INVALID!': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'waste combustion (community)': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'LPG special condition': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'appliances able to use mineral oil or liquid biofuel': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'electricity - this is for backwards compatibility only and should not be used': {'price': None, 'primary_factor': None, 'emission_factor': None},
    'biomass - this is for backwards compatibility only and should not be used': {'price': None, 'primary_factor': None, 'emission_factor': None}
}
hot_water_data = {
    'Community scheme': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, off-peak': {'price': 0.0661, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'No system present: electric immersion assumed': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Electric immersion, standard tariff': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Electric instantaneous at point of use': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'From main system, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas multipoint': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, no cylinder thermostat, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil range cooker': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'Community scheme, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, no cylinderstat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas boiler/circulator, no cylinder thermostat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil range cooker, no cylinder thermostat': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'community scheme': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas range cooker, no cylinder thermostat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, plus solar, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas boiler/circulator': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas instantaneous at point of use': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, off-peak, plus solar': {'price': 0.0661, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Electric immersion, standard tariff, plus solar': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'From secondary system, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil boiler/circulator, no cylinder thermostat': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'From main system, no cylinderstat, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, off-peak, no cylinder thermostat': {'price': 0.0661, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Oil boiler/circulator': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'community scheme, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From secondary system': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Solid fuel boiler/circulator': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'From main system, waste water heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Community scheme with CHP': {'price': 0.0297, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Community scheme, no cylinder thermostat, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, standard tariff, no cylinder thermostat': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Gas range cooker': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, flue gas heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Community scheme, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Solid fuel boiler/circulator, no cylinder thermostat': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'Solid fuel range cooker, no cylinder thermostat': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'From secondary system, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas range cooker, no cylinderstat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'No system present: electric immersion assumed, plus solar': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Electric instantaneous at point of use, plus solar': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Solid fuel range cooker, no cylinderstat': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'Solid fuel range cooker, plus solar, no cylinder thermostat': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'Solid fuel range cooker': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'No hot water system present - electric immersion assumed, plus solar': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'No hot water system present - electric immersion assumed': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    "O'r brif system": {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From secondary system, no cylinderstat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'No system present?electric immersion assumed': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'O?r brif system, dim thermostat ar y silindr': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, no cylinder thermostat, waste water heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas boiler/circulator, no cylinderstat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Solid fuel range cooker, plus solar': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'Electric instantaneous at point of use, no cylinder thermostat': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'From main system, plus solar, flue gas heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From second main heating system, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, off-peak, waste water heat recovery': {'price': 0.0661, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Community heat pump': {'price': 0.0297, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Gas multipoint, plus solar': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil boiler/circulator, plus solar': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'Gas boiler/circulator, plus solar': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil range cooker, plus solar': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'Solid fuel boiler/circulator, plus solar': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'Gas boiler/circulator, no cylinderstat, plus solar': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From community scheme, plus solar': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil range cooker, no cylinderstat': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'No system present: electric immersion assumed, flue gas heat recovery': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Solid fuel boiler/circulator, plus solar, no cylinder thermostat': {'price': 0.0367, 'primary_factor': 1.00, 'emission_factor': 0.394},
    'O?r brif system': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, no cylinder thermostat, flue gas heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From secondary system, plus solar, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From second main heating system': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Community scheme, plus solar, no cylinder thermostat': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas range cooker, plus solar': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Twymwr tanddwr, an-frig': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Dim system ar gael: rhagdybir bod twymwr tanddwr trydan': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Community scheme with CHP, plus solar': {'price': 0.0297, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'Oil range cooker, plus solar, no cylinder thermostat': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'Community scheme with CHP, no cylinder thermostat': {'price': 0.0297, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'No system present?electric immersion assumed, no cylinder thermostat': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519},
    'SAP:Hot-Water': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas multipoint, no cylinder thermostat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'From main system, waste water heat recovery, flue gas heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Oil range cooker, no cylinderstat, plus solar': {'price': 0.0544, 'primary_factor': 1.10, 'emission_factor': 0.298},
    'Gas range cooker, plus solar, no cylinder thermostat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Gas instantaneous at point of use, no cylinder thermostat': {'price': 0.0348, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Community scheme, waste water heat recovery': {'price': 0.0424, 'primary_factor': 1.22, 'emission_factor': 0.216},
    'Electric immersion, standard tariff, waste water heat recovery': {'price': 0.1319, 'primary_factor': 3.07, 'emission_factor': 0.519}
}

# Function to get fuel data
def get_fuel_data(fuel_type, data_type):
    return fuel_data.get(fuel_type, {}).get(data_type, None)

def get_hot_water_data(hot_water_type, data_type):
    return hot_water_data.get(hot_water_type, {}).get(data_type, None)

# Map the dictionaries to the dataset
df['Fuel_Price'] = df['MAIN_FUEL'].apply(lambda x: get_fuel_data(x, 'price'))
df['Primary_Energy_Conversion_Factor'] = df['MAIN_FUEL'].apply(lambda x: get_fuel_data(x, 'primary_factor'))
df['Emission_Factor'] = df['MAIN_FUEL'].apply(lambda x: get_fuel_data(x, 'emission_factor'))

df['Hot_Water_Price'] = df['HOTWATER_DESCRIPTION'].apply(lambda x: get_hot_water_data(x, 'price'))
df['Hot_Water_Primary_Energy_Conversion_Factor'] = df['HOTWATER_DESCRIPTION'].apply(lambda x: get_hot_water_data(x, 'primary_factor'))
df['Hot_Water_Emission_Factor'] = df['HOTWATER_DESCRIPTION'].apply(lambda x: get_hot_water_data(x, 'emission_factor'))

# Calculate Heating consumption and intensity
df['Heating_Cost'] = df['HEATING_COST_CURRENT']  # Assuming HEATING_COST_CURRENT is already calculated
df['Annual_Energy_Consumption_Heating'] = df['Heating_Cost'] / df['Fuel_Price']
# Handle zero or missing values in TOTAL_FLOOR_AREA
df['TOTAL_FLOOR_AREA'].replace(0, np.nan, inplace=True)
# Calculate Final Energy Intensity for heating, hot water, and lighting
df['Heating_Final_Energy_Intensity'] = df['Annual_Energy_Consumption_Heating'] / df['TOTAL_FLOOR_AREA']
df['Heating_Primary_Energy_Intensity'] = df['Heating_Final_Energy_Intensity'] * df['Primary_Energy_Conversion_Factor']

# Calculate Hot Water consumption and intensity
df['Hot_Water_Cost'] = df['HOT_WATER_COST_CURRENT']
df['Annual_Energy_Consumption_Hot_Water'] = df['Hot_Water_Cost'] / df['Hot_Water_Price']
df['Hot_Water_Final_Energy_Intensity'] = df['Annual_Energy_Consumption_Hot_Water'] / df['TOTAL_FLOOR_AREA']
df['Hot_Water_Primary_Energy_Intensity'] = df['Hot_Water_Final_Energy_Intensity'] * df['Hot_Water_Primary_Energy_Conversion_Factor']

# Calculate Lighting consumption and intensity
df['Lighting_Cost'] = df['LIGHTING_COST_CURRENT']
df['Annual_Energy_Consumption_Lighting'] = df['Lighting_Cost'] / 0.1319
df['Lighting_Final_Energy_Intensity'] = df['Annual_Energy_Consumption_Lighting'] / df['TOTAL_FLOOR_AREA']
df['Lighting_Primary_Energy_Intensity'] = df['Lighting_Final_Energy_Intensity'] * 3.07

# Sum all primary energy intensities
df['Total_Primary_Energy_Intensity'] = df['Heating_Primary_Energy_Intensity'] + df['Hot_Water_Primary_Energy_Intensity'] + df['Lighting_Primary_Energy_Intensity']

# Define the deflator constant (as per SAP methodology)
deflator = 0.42

# Calculate the Energy Cost Factor (ECF)
df['ECF'] = deflator * (df['Heating_Cost']+df['Hot_Water_Cost']+df['Lighting_Cost']) / (df['TOTAL_FLOOR_AREA'] + 45)

# Calculate the SAP Rating based on ECF
def calculate_sap_rating(ecf):
    if ecf >= 3.5:
        return 117 - 121 * np.log10(ecf)
    else:
        return 100 - 13.95 * ecf

df['SAP_Rating'] = df['ECF'].apply(calculate_sap_rating)

# Round the SAP Rating to the nearest integer and handle NaN values
df['SAP_Rating'] = df['SAP_Rating'].fillna(0).round().astype(int)

# Determine EPC Band
def determine_epc_band(sap_score):
    if sap_score >= 92:
        return 'A'
    elif sap_score >= 81:
        return 'B'
    elif sap_score >= 69:
        return 'C'
    elif sap_score >= 55:
        return 'D'
    elif sap_score >= 39:
        return 'E'
    elif sap_score >= 21:
        return 'F'
    else:
        return 'G'

df['EPC_Band'] = df['SAP_Rating'].apply(determine_epc_band)

# Display the results
print(df[['MAIN_FUEL', 'Fuel_Price', 'Heating_Cost', 'Hot_Water_Cost', 'Lighting_Cost', 'Annual_Energy_Consumption_Heating', 
          'Annual_Energy_Consumption_Hot_Water', 'Annual_Energy_Consumption_Lighting', 'Heating_Final_Energy_Intensity',
          'Hot_Water_Final_Energy_Intensity', 'Lighting_Final_Energy_Intensity', 'Heating_Primary_Energy_Intensity',
          'Hot_Water_Primary_Energy_Intensity', 'Lighting_Primary_Energy_Intensity', 'Total_Primary_Energy_Intensity', 
          'SAP_Rating', 'EPC_Band']])
```

```python
# Calculate CO2 EMISS CURR PER FLOOR AREA [kgCo2eq/m2/year] from Final energy intensities and pertinent emission factors [kgCo2eq/kWh] from the methodology

# Calculate CO2 emissions for each usage
df['Heating_CO2_Emissions'] = df['Heating_Final_Energy_Intensity'] * df['Emission_Factor']
df['Hot_Water_CO2_Emissions'] = df['Hot_Water_Final_Energy_Intensity'] * df['Hot_Water_Emission_Factor']
df['Lighting_CO2_Emissions'] = df['Lighting_Final_Energy_Intensity'] * 0.519  # Using the emission factor for electricity

# Sum the CO2 emissions to get the total CO2 emissions per floor area
df['Total_CO2_Emissions_per_Floor_Area'] = df['Heating_CO2_Emissions'] + df['Hot_Water_CO2_Emissions'] + df['Lighting_CO2_Emissions']

# Display the CO2 emissions results
print(df[['Heating_CO2_Emissions', 'Hot_Water_CO2_Emissions', 'Lighting_CO2_Emissions', 'Total_CO2_Emissions_per_Floor_Area']].head())
```
